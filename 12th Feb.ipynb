```python
#Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.

#Ans, Custom Exceptions are a class derived from built in Exception class. We need custom Exception as makes our program  easier to follow and more readable as an exception sholdn't be complicated.

#EXAMPLE:
class Contacterror(TypeError):
    def __init__(self, arg):
        self.args = arg
 
try:
    raise Contacterror("Error")
 
except Contacterror as e:
    print(e.args)
```
