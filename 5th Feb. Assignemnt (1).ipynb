```python
#Q5. What is inheritance? Give an example for each type of inheritance.
#Ans. Inheritance method is the method of inheriting the propertise of parent class into a child class.

#EXAMPLE:
class Employee:

    def name(self):
        print("This function is in Employee class.")

class Location(Employee):

    def city(self):
        print("This function is in Location class.")

object = Location()

object.name()
object.city()

```

```python
#Below are examples for each type of inheritance

#Example of Single Inheritance
class Employee:

    def name(self):
        print("This function is in Employee class.")

class Location(Employee):

    def city(self):
        print("This function is in Location class.")

object = Location()

object.name()
object.city()
```

```python
#Example of Multiple Inheritance
class Employee:

    Employee_name = ""

    def Employee(self):

        print(self.Employee_name)

class Location:

    Location_name = ""

    def Location(self):

        print(self.Location_name)

class Department(Employee, Location):

    def MIS(self):

        print("Father :", self.Employee_name)

        print("Mother :", self.Location_name)

s1 = Department()

s1.Employee_name = "SAHIL"

s1.Location_name = "Bhopal"

s1.MIS()
```

```python
#Example of Multilevel Inheritance

class Grandfather:

    def __init__(self, grandfathername):

        self.grandfathername = grandfathername

class Father(Grandfather):

    def __init__(self, fathername, grandfathername):

        self.fathername = fathername

        Grandfather.__init__(self, grandfathername)

class Son(Father):

    def __init__(self, sonname, fathername, grandfathername):

        self.sonname = sonname

        Father.__init__(self, fathername, grandfathername)

    def print_name(self):

        print('Grandfather name :', self.grandfathername)

        print("Father name :", self.fathername)

        print("Son name :", self.sonname)

s1 = Son('Joseph', 'Filomina', 'ALexzender')
print(s1.grandfathername)
s1.print_name()

```

```python
#Example of Hierarchical Inheritance

class Parent:

    def func1(self):

        print("This function is in parent class.")

class Child1(Parent):

    def func2(self):

        print("This function is in child 1.")

class Child2(Parent):

    def func3(self):

        print("This function is in child 2.")

object1 = Child1()

object2 = Child2()

object1.func1()

object1.func2()

object2.func1()

object2.func3()


```

```python
#Example of Hybrid Inheritance
class School:

    def func1(self):

        print("This function is in school.")

class Student1(School):

    def func2(self):

        print("This function is in student 1. ")

class Student2(School):

    def func3(self):

        print("This function is in student 2.")

class Student3(Student1, School):

    def func4(self):

        print("This function is in student 3.")

object = Student3()

object.func1()

object.func2()
```
